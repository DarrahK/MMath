Plot of different second order parametric models 

**NOTE**: This code is just for experimentation/explanation purposes for my thesis only and is not intended to be a demonstration of my coding ability.

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
def poly1(x):
  return 1 + 2 * x + 4 * x**2
```

```python
def poly2(x):
  return 4 + 2 * x + 10 * x**2
```

```python
def poly3(x):
  return 3 - 10 * x - 3 * x**2
```

```python
x_plot = np.linspace(0.0, 1.0, 100)

figure(figsize=(6,6))
plt.plot(x_plot, poly1(x_plot), label = "f_θ(x) = 1 - 2x + 4x^2")
plt.plot(x_plot, poly2(x_plot), label = "f_θ(x) = 4 + 2x + 10x^2")
plt.plot(x_plot, poly3(x_plot), label = "f_θ(x) = 3 - 10x - 3x^2")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.title("Plot of different second order parametric models f_θ")
plt.legend(loc="lower left")
plt.show()
```
