**NOTE**: This code is just for experimentation/explanation purposes for my thesis only and is not intended to be a demonstration of my coding ability.

```python
import matplotlib.pyplot as plt
import numpy as np

sigma = 0.1
mu = 0.5
s = np.random.normal(mu, sigma, 1000)

# Create the bins and histogram
count, bins, ignored = plt.hist(s, 20, density = True)

# Plot the distribution curves
mu = 0.4
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, label = 'θ = 0.4')

mu = 0.5
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color = "y", label = 'θ = 0.5')

mu = 0.6
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, label = 'θ = 0.6')

plt.legend()
plt.title("Fitting the PDF of N(θ, 0.1) to N(0.5, 0.1)")
plt.show()
```
